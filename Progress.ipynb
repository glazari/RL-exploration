```python
def mean(array, window=2):
    n = window
    l = len(array)
    return [np.mean(array[i:i+n]) for i in range(l-n+1)]
```

```python
def maxi(array, window=2):
    n = window
    l = len(array)
    return [np.max(array[i:i+n]) for i in range(l-n+1)]
```
