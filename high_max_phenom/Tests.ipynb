```python
import os
import sys
import pickle
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

sys.path.append('..')
from plot_assist import plot_compare, display_videos
```

```python
folders = [
    'Test1/',
    'Test2/',
    'Seaquest_explorer/',
    'Seaquest_2_explorer/',
    'Seaquest_3_explorer/',
]

all_stats = []
for folder in folders:
    folder_files = os.listdir(folder)
    stats_file = [name for name in folder_files if 'stats' in name][0]
    
    with open(folder+stats_file,'rb') as f:
        stats = pickle.load(f)
    
    all_stats.append(stats)
    

life_stats = [stats['life episode'] for stats in all_stats]
epi_stats = [stats['real episode'] for stats in all_stats]
```

# Life episodes

### Clipped Rewards

```python
series = [stats['clipped reward'] for stats in life_stats]
max_series = max([max(ser) for ser in series])*1.05
scale = [0, 18000, 0, max_series]
titles = ['test '+str(i+1) for i in range(len(series))]

plot_compare(series=series, scale=scale, titles=titles)
```
