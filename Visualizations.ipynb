```python
import numpy as np
import matplotlib.pyplot as plt
import gym
import os
%matplotlib inline
```

```python
action_files = os.listdir('Graphics/actions/')

action_images = {}
for action_file in action_files:
    image =plt.imread('Graphics/actions/'+action_file)
    name = action_file.replace('.bmp','')
    action_images[name] = image
```

```python
env = gym.make('Seaquest-v4')
obs = env.reset()
```

```python
action_names = env.unwrapped.get_action_meanings()
```

```python
def obs_action(action, obs):
    act_name = action_names[action]
    act_img = action_images[act_name]
    plt.imshow(np.concatenate((obs,act_img),axis=1))
    plt.show()
```

```python
done = False
for i in range(10):
    action = env.action_space.sample()
    obs, rew, done, info = env.step(action)
    obs_action(action,obs)
```
